```python
import numpy as np
import pandas as pd
import pantab
```

```python
df_norm = pd.read_csv('sts_norm.csv', sep=',')
df_collar = pd.read_csv('dwh_collar.csv', sep = ',')
```

```python
df_norm.head()
```

```python
df1 = df_norm[(df_norm['normalization_model'] == 'INFOJOBS-EXPERIENCE_JOBTITLE-1.3.0')
              | (df_norm['normalization_model'] == 'INFOJOBS-OFFER_JOBTITLE-1.3.0'          
        )]
```

```python
df1['DESC_PUESTO_NORM'] = df1['normalized_value']
```

```python
df_collar.head()
```

```python
df_merge_collar = df1.merge(df_collar, how='left', on='DESC_PUESTO_NORM')
```

```python
list(df_merge_collar.columns)
```

```python
df_merge_collar_2= df_merge_collar[['normalization_model',
 'type',
 'normalized_value',
 'is_known',
 'num_rows',
 'correct_prediction',
 'score',
 'dt',
 'N_NORMALIZATION_ID',
 'ID_4_x',
 'DESC_PUESTO_NORM',
 'ID',
 'COLLAR1',
 'COLLAR2']]
```

```python
df_merge_collar_2.to_csv('df_merge_collar_2.csv')
```

```python
df_merge_collar_2_nojoin=df_merge_collar_2[df_merge_collar_2['COLLAR1'].isna()]
```

```python
df_merge_collar_2_nojoin
```

```python
df_merge_collar_2_nojoin.to_csv('df_merge_collar_2_nojoin.csv')
```
